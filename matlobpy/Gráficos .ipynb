```python
import matplotlib.pyplot as plt
import numpy as np
fig = plt.figure() #Um container que pode conter vários gráficos
ax = plt.axes() #Um container especifico dentro da figure que representa um gráfico

#axis são os eixos

x = np.linspace(0, np.pi*2.5) #gera os valores de 0 a 7
ax.plot(x,np.sin(x)) #plota o gráfico com x valores e y sendo o seno de x
plt.show()# mostra o gráfico
```

```python
#utilizando subplots
#axis são os eixos
#subplots OO
x = np.linspace(0, np.pi*3) #gera os valores de 0 a 7

fig,ax = plt.subplots(nrows=2) #cria uma área de gráfico com 2 linhas
fig.axes[0].plot(x,np.sin(x),label = "seno")

fig.axes[1].plot(x,np.cos(x),label = "cosseno")

fig.axes[1].legend(loc="best")

```

```python
x = np.linspace(0, np.pi*2.25)
fig = plt.figure()
ax = plt.axes()
ax.set_title("Titulo do plot")
ax.set_xlabel("x")
ax.set_ylabel("sen(x)")
ax.plot(x, np.sin(x),linestyle='--',color='r')

```

```python
#estilo matlab
x = np.linspace(0, np.pi*2.25)

plt.plot(x,np.sin(x))
```

```python
#Subplost estilo matlab
x = np.linspace(0, np.pi*2.10)
plt.subplot(2,1,1) #2 linhas 1 colunas 1 linha 1 a atual
plt.plot(x,np.sin(x))
plt.subplot(2,1,2) #2 linhas 1 coluna e linha 2 a atual
plt.plot(x,np.cos(x))
```

```python
#Titulos e rótulos no estilo MATLAB
x = np.linspace(0, np.pi*2.25)
plt.title("Titulo do plot")
plt.xlabel("X")
plt.ylabel("sen(x)")
plt.plot(x,np.sin(x))
```

```python
#Titulo e rótulos OO
x = np.linspace(0, np.pi*2.25)
fig = plt.figure()
ax = plt.axes()
ax.set_title("Titulo do plot")
ax.set_xlabel("x")
ax.set_ylabel("sen(x)")
ax.plot(x,np.sin(x),linestyle = "dashdot",color="g")
```

```python
x = np.linspace(0, np.pi*2.25)
plt.title("Exemplo com legenda")
plt.plot(x,np.sin(x),label="seno")
plt.plot(x,np.cos(x),label="cosseno")
plt.legend(loc="lower right")
```

```python
x = np.linspace(0, np.pi*2.25)
ax = plt.axes()
ax.plot(x,np.sin(x))

ax.set(xlim=(0,np.pi*2),
      ylim=(-1.25,1.25),
       xlabel = "x",
       ylabel = "sin(x)",
       title = "Usando ax.set()"
      );
```
